```python
%%writefile app.py

import streamlit as st
import pandas as pd
import plotly.express as px

# ====================================
# PAGE CONFIG
# ====================================

st.set_page_config(
    page_title="AI Banking Insight Dashboard",
    layout="wide"
)

# ====================================
# TITLE
# ====================================

st.title("🏦 AI Banking Insight Generator")

st.markdown("""
This AI Dashboard Automatically:
- Analyzes Banking Data
- Generates AI Insights
- Detects Financial Risks
- Suggests Business Solutions
""")

# ====================================
# LOAD DATA
# ====================================

df = pd.read_excel("Advanced_Cleaned_Messy.xlsx")

# Clean column names
df.columns = df.columns.str.lower().str.strip()

# ====================================
# KPI SECTION
# ====================================

total_customers = len(df)

avg_income = df['estimated_income'].mean()

total_loans = df['bank_loans'].sum()

total_deposits = df['bank_deposits'].sum()

# KPI Cards
c1, c2, c3, c4 = st.columns(4)

c1.metric("Total Customers", f"{total_customers:,}")

c2.metric("Average Income", f"{avg_income:,.0f}")

c3.metric("Total Loans", f"{total_loans:,.0f}")

c4.metric("Total Deposits", f"{total_deposits:,.0f}")

# ====================================
# AI GENERATED INSIGHTS
# ====================================

st.header("🤖 AI Generated Insights")

# Insight 1
if avg_income > 100000:

    st.success("""
    Insight:
    Customers have strong income levels.

    Business Meaning:
    Customers are financially stable.

    Solution:
    Launch premium banking and investment plans.
    """)

# Insight 2
risky = df[df['bank_loans'] > df['bank_deposits']]

if len(risky) > 0:

    st.warning(f"""
    Insight:
    {len(risky)} customers have loans higher than deposits.

    Problem:
    Possible repayment risk.

    Solution:
    Improve loan monitoring systems.
    """)

# Insight 3
high_cc = df[df['credit_card_balance'] > 50000]

if len(high_cc) > 0:

    st.info(f"""
    Insight:
    {len(high_cc)} customers show high credit card usage.

    Solution:
    Offer EMI and reward systems.
    """)

# Insight 4
senior = df[df['age'] > 55]

if len(senior) > 0:

    st.success(f"""
    Insight:
    {len(senior)} senior customers maintain stable savings.

    Solution:
    Promote retirement banking products.
    """)

# ====================================
# CHARTS
# ====================================

st.header("📊 Banking Analytics")

# Gender Chart
gender = df['genderid'].value_counts().reset_index()

gender.columns = ['Gender', 'Count']

fig1 = px.bar(
    gender,
    x='Gender',
    y='Count',
    title='Gender Distribution'
)

st.plotly_chart(fig1, use_container_width=True)

# Occupation Chart
occ = df['occupation'].value_counts().head(10).reset_index()

occ.columns = ['Occupation', 'Count']

fig2 = px.bar(
    occ,
    x='Occupation',
    y='Count',
    title='Top Occupations'
)

st.plotly_chart(fig2, use_container_width=True)

# Loan vs Deposit Chart
finance = pd.DataFrame({
    'Category': ['Loans', 'Deposits'],
    'Amount': [total_loans, total_deposits]
})

fig3 = px.pie(
    finance,
    names='Category',
    values='Amount',
    title='Loans vs Deposits'
)

st.plotly_chart(fig3, use_container_width=True)

# Age Distribution
fig4 = px.histogram(
    df,
    x='age',
    nbins=20,
    title='Customer Age Distribution'
)

st.plotly_chart(fig4, use_container_width=True)

# ====================================
# HIGH RISK CUSTOMERS
# ====================================

st.header("⚠️ High Risk Customers")

risk_df = df[df['bank_loans'] > df['bank_deposits']]

st.dataframe(
    risk_df[
        ['name', 'bank_loans', 'bank_deposits']
    ].head(20)
)

# ====================================
# DATA PREVIEW
# ====================================

st.header("📄 Dataset Preview")

st.dataframe(df.head())

# ====================================
# FINAL SUMMARY
# ====================================

st.header("📌 Final AI Summary")

st.info("""
This AI Banking Dashboard:
- Detects financial risks
- Generates business insights
- Analyzes customer behavior
- Suggests business solutions
- Creates interactive analytics
""")

# ====================================
# FOOTER
# ====================================

st.markdown("---")
st.markdown("Developed using Python, Streamlit, AI Analytics")
```

```python
# ================================
# AI BANKING INSIGHT DASHBOARD
# FULL SINGLE FILE PROJECT
# ================================

# INSTALL FIRST:
# pip install streamlit pandas plotly google-generativeai openpyxl

import streamlit as st
import pandas as pd
import plotly.express as px
import google.generativeai as genai

# ================================
# PAGE CONFIG
# ================================

st.set_page_config(
    page_title="AI Banking Dashboard",
    page_icon="🤖",
    layout="wide"
)

# ================================
# GEMINI API
# ================================

GOOGLE_API_KEY = "YOUR_GEMINI_API_KEY"

genai.configure(api_key=GOOGLE_API_KEY)

model = genai.GenerativeModel("gemini-1.5-flash")

# ================================
# LOAD DATA
# ================================

@st.cache_data
def load_data():
    df = pd.read_excel("Advanced_Cleaned_Messy(6).xlsx")
    return df

df = load_data()

# ================================
# SIDEBAR
# ================================

st.sidebar.title("🤖 AI Banking Dashboard")

menu = st.sidebar.radio(
    "Navigation",
    [
        "Dashboard",
        "AI Insights",
        "Client Analysis",
        "Revenue Analysis",
        "Custom Query"
    ]
)

# ================================
# TOP HEADER
# ================================

st.title("🏦 AI Banking Insight Dashboard")

st.markdown("### Ask Anything About Your Dataset")

# ================================
# SEARCH BOX
# ================================

user_question = st.text_input(
    "🔍 Ask Your Question",
    placeholder="Example: Show top occupations or Client ID details"
)

# ================================
# KPI CARDS
# ================================

col1, col2, col3, col4 = st.columns(4)

with col1:
    st.metric("Total Clients", len(df))

with col2:
    if "Revenue" in df.columns:
        st.metric("Total Revenue", f"{df['Revenue'].sum():,.0f}")

with col3:
    if "Occupation" in df.columns:
        st.metric(
            "Top Occupation",
            df["Occupation"].mode()[0]
        )

with col4:
    if "Nationality" in df.columns:
        st.metric(
            "Total Nationality",
            df["Nationality"].nunique()
        )

# ================================
# DASHBOARD
# ================================

if menu == "Dashboard":

    st.subheader("📊 Dataset Preview")

    st.dataframe(df.head())

    # OCCUPATION CHART
    if "Occupation" in df.columns:

        occ = df["Occupation"].value_counts().head(10)

        fig = px.bar(
            occ,
            x=occ.index,
            y=occ.values,
            title="Top Occupations"
        )

        st.plotly_chart(fig, use_container_width=True)

    # GENDER CHART
    if "Gender" in df.columns:

        gender = df["Gender"].value_counts()

        fig2 = px.pie(
            names=gender.index,
            values=gender.values,
            title="Gender Distribution"
        )

        st.plotly_chart(fig2, use_container_width=True)

# ================================
# CLIENT ANALYSIS
# ================================

elif menu == "Client Analysis":

    st.subheader("👤 Client Analysis")

    if "Client ID" in df.columns:

        client_id = st.selectbox(
            "Select Client ID",
            df["Client ID"].unique()
        )

        client_data = df[df["Client ID"] == client_id]

        st.dataframe(client_data)

# ================================
# REVENUE ANALYSIS
# ================================

elif menu == "Revenue Analysis":

    st.subheader("💰 Revenue Analysis")

    if "Revenue" in df.columns and "Occupation" in df.columns:

        revenue_data = df.groupby("Occupation")["Revenue"].sum().reset_index()

        fig3 = px.bar(
            revenue_data,
            x="Occupation",
            y="Revenue",
            title="Revenue by Occupation"
        )

        st.plotly_chart(fig3, use_container_width=True)

# ================================
# AI INSIGHTS
# ================================

elif menu == "AI Insights":

    st.subheader("🧠 AI Generated Insights")

    prompt = f"""
    Analyze this banking dataset.

    Columns:
    {df.columns.tolist()}

    Generate:
    1. Key Insights
    2. Important Trends
    3. Business Recommendations
    4. Client Behavior Analysis
    """

    response = model.generate_content(prompt)

    st.write(response.text)

# ================================
# CUSTOM QUERY
# ================================

elif menu == "Custom Query":

    st.subheader("🤖 Chat With Your Data")

    if user_question:

        # SIMPLE QUERY SYSTEM

        question = user_question.lower()

        # TOP OCCUPATION
        if "top occupation" in question:

            result = df["Occupation"].value_counts().head(10)

            st.write(result)

            fig4 = px.bar(
                result,
                x=result.index,
                y=result.values,
                title="Top Occupations"
            )

            st.plotly_chart(fig4, use_container_width=True)

        # CLIENT ID DETAILS
        elif "client id" in question:

            numbers = ''.join(filter(str.isdigit, question))

            if numbers:

                client_data = df[df["Client ID"].astype(str) == numbers]

                st.dataframe(client_data)

            else:
                st.warning("Please enter valid Client ID")

        # HIGHEST REVENUE
        elif "highest revenue" in question:

            if "Revenue" in df.columns:

                max_rev = df["Revenue"].max()

                row = df[df["Revenue"] == max_rev]

                st.write("Highest Revenue Record")

                st.dataframe(row)

        # AI GENERAL RESPONSE
        else:

            prompt = f"""
            You are an AI Data Analyst.

            Dataset Columns:
            {df.columns.tolist()}

            User Question:
            {user_question}

            Give detailed answer from banking dataset.
            """

            response = model.generate_content(prompt)

            st.write(response.text)

# ================================
# FOOTER
# ================================

st.markdown("---")
st.markdown("🚀 AI Powered Banking Analytics Dashboard")
```
